##In this notebook: 
###Choose and sample outliers with interactive functions.

The following functions are used:

  * `study.interactive_pca`
     * This funciton performs [Principal Components Ananlysis](http://scikit-learn.org/stable/modules/decomposition.html) on user-selected subsets of samples and features. Samples are colored according to a dictionary saved in the `Create A New Study` notebook. This function will be examined in more depth in the [Dimensionality Reduction](3_Dimensionality%20Reduction.ipynb) notebook.
  * `study.interactive_choose_outliers`
     * This function uses a radial basis function and one-class SVM to identify outliers based on user-chosen fitting parameters.
  * `study.interactive_reset_outliers`
     * User can combine multiple outlier calls to set `study.metadata.data['outlier']`
  * `study.drop_outliers`
     * After setting the `outlier` attribute in `study.metadata.data` the samples labelled outliers are removed from `study.expression.data` and `study.splicing.data`
